```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import pandas as pd
import math


mu = 15.7

A = np.array([11.5, 9.8, 10.7, 16.1, 14.1, 10.5, 15.2, 19.0, 12.8, 12.4, 19.2, 13.5])
s = A.std()

df = 11

ttest = stats.ttest_1samp(A, mu)

tcrit = -1.796


print "t-statistic: ", round(ttest.statistic, 5)
print "p-value: ", ttest.pvalue

# ig, ax = plt.subplots(1, 1)

rv = stats.t(df=11, loc=0, scale=1)
x = np.linspace(rv.ppf(0.001), rv.ppf(0.999), 100)
y = rv.pdf(x) 
plt.xlim(-4,4)
plt.plot(x,y,lw=3, alpha=0.5, color = 'purple')
         
# x = np.linspace(stats.t.ppf(0.01, df),stats.t.ppf(0.99, df), 100)
# ax.plot(x, stats.t.pdf(x, df),'r-', lw=4, alpha=0.7, color = 'purple')
# ax.set_xlim(-2.6, 2.6)

plt.axvline(x=ttest.statistic,ymax=0.14, linestyle='--', label="t-statistic=%.3f" %ttest.statistic)
x_tcrit = plt.axvline(x=tcrit,ymax=0.23, color = 'r', label="t-critical= -1.796")

# labels
plt.title("Student's t probability density distribution for df=11")
plt.xlabel('x')
plt.ylabel('P(x)')

# plt.xticks([ttest.statistic,tcrit], [round(ttest.statistic,3),tcrit],  rotation=20)

plt.legend(loc = 1)

plt.show()
```

```python
# roll a dice 20 times, sum their outcomes 
# repeat experiment 1000 times and plot the distribution of results

nExps = 1000
nDice = 20


diceRolls=scipy.random.randint(6,size=(nExps,nDice))+1

# compute the mean
M = np.sum(diceRolls,1)

# plot all estimated mean
plt.hist(M,normed=True)

x = np.linspace(np.min(M),np.max(M), 100)

plt.plot(x,stats.norm(np.mean(M),np.std(M)).pdf(x),'r',lw=3)

print("Fair fit with the normal distribution")
```

**Taking several samples to estimate the mean**

```python
# roll a dice 100 times to estimate the its mean value (note, it should be 3.5)
# repeat experiment 1000 times

nExps = 1000
nDice = 2

diceRolls=scipy.random.randint(6,size=(nExps,nDice))+1

# compute the mean
M = np.mean(diceRolls,1)

# plot all estimated mean
plt.hist(M,normed=True)

plt.xlim([1,6])

x = np.linspace(np.min(M),np.max(M), 100)

plt.plot(x,stats.norm(np.mean(M),np.std(M)).pdf(x),'r',lw=3)

print("High dispersion of estimates")
```
